```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#s
mu = 0.1
#nbr of steps
n = 100
#time in year
T = 1
#nbr of sima
M = 4
#initial stock price
S0 = 100
#volatilite
sigma = 0.3
```

```python
#calcule each step
dt = T/n
#simulating using np array
St = np.exp(
    (mu - sigma ** 2 / 2 ) * dt
    + sigma * np.random.normal(0, np.sqrt(dt), size = (M,n)).T
)
#imclude array of ones
St = np.vstack([np.ones(M), St])

#multiply bu S0 
St = S0 * St.cumprod(axis=0)
```

```python
time = np.linspace(0, T, n+1)


tt = np.full(shape=(M, n+1), fill_value=time).T
```

```python
plt.plot(tt, St)
plt.show()
```
